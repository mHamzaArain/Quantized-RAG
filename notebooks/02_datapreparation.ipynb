```python
# Extract necessary info only
data = []

for unprocessed_text in dataset['test_10_percent']['text'][:]:
    a = unprocessed_text.split('<s>[INST] ')
    b = a[1].split('[/INST] ')
    datum = b[1].split(' </s>')[0]
    datum = datum
    data.append(datum)

print(len(data))

# Remove redundancy
data = list(set(data))
len(data)
```

```python
data
```

```python
def text_length_statistics(text_list):
    # Calculate lengths of all texts in the list
    lengths = [len(text) for text in text_list]
    
    # Calculate minimum length
    min_length = min(lengths)
    
    # Calculate maximum length
    max_length = max(lengths)
    
    # Calculate average length
    avg_length = sum(lengths) / len(lengths) if lengths else 0
    
    return min_length, max_length, avg_length

min_len, max_len, avg_len = text_length_statistics(data)
print(f"Minimum length: {min_len}")
print(f"Maximum length: {max_len}")
print(f"Average length: {avg_len:.2f}")
```
